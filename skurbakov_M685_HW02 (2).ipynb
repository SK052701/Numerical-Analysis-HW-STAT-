```python
def polynomial(x):
  y = x**3+x-1
  return y


#bisection root finding

def bisection(aa,ab, aepsilon):
  # biiig assumption that f(a)*f(b) < 0
  a = aa;b = ab # habit to save args if changing them
  fa = polynomial(a)
  isteps = 0

  while abs((b-a)/2.) > aepsilon:
    c = (a+b)/2.
    fc = polynomial(c)
    isteps += 1 # common shortcut for isteps = isteps + 1
    if fc == 0:
      break
    elif polynomial(a)*fc < 0:
      b = c
    else:
      a = c
  return c,isteps
```

```python
print(bisection(0,1,0.00001))
print(bisection(0,1,0.000001))
```

```python
#newtons root finding

def f_newton(af, adf, ax,aepsilon,aimax):
  root = 0;iterations = 0

  icount = 0
  x_old = ax
  fx = af(x_old)
  while (abs(fx) > aepsilon) and (icount < aimax): ##
    #x_new = x_old - f(x_old)/f'(x_old)
    x_new = x_old - fx/adf(x_old) # at this point, fx = f(x_old) ~ fx_old
    #fx = f(x_new)
    fx = af(x_new)
    #x_old = x_new
    x_old = x_new
    #icount = icount + 1
    icount += 1

  root = x_old;iterations = icount
  return root,iterations

def df(ax):
  rc = 3*ax**2 + 1 # f'(x) = d(x^3 + x - 1)
  return rc
```

```python
print(f_newton(polynomial,df,0.3, 0.00001, 50))
print(f_newton(polynomial,df,0.3, 0.000001, 50))
```

```python
#secant root finding method
def qf(af,ax1,ax2):
  # difference quotient here, plz
  rc = (af(ax2)-af(ax1))/(ax2-ax1)
  return rc

def f_secant(af,ax1,ax2,aepsilon,aimax):
  root = 0;iterations = 0

  # do some stuffs here
  icount = 0
  x1,x2 = ax1, ax2
  fx = af(x2)
  while (abs(fx) > aepsilon) and (icount < aimax): ##
    x_new = x2 - fx/qf(af,x1,x2)
    #fx = f(x_new)
    fx = af(x_new)
    #x_old = x_new
    x1, x2 = x2, x_new
    #icount = icount + 1
    icount += 1

  root = x2;iterations = icount
  return root,iterations

```

```python
print(f_secant(polynomial, 0.1, 0.9, 0.00001, 50))
print(f_secant(polynomial, 0.1, 0.9, 0.000001, 50))
```

The Newton and the secant method use the same overall algorithm for root finding. The only difference is that Newton's method uses the tangent line while the secant method uses the secant line. Because of of that Newton's method requires only 1 point in the domain to be guessed to start the algorithm, while the secant method requires 2 points in the domain to be guessed.
Normally the Newton method ususally converges faster than the secant method. Newton method converges quadratically, while the secant method linearly. This means that with each iteration the error between the approximation and the actual solution decreases at a contstant rate for secant method vs decreasing at a rate that's proportional to the squared error from the previous iteration for newton's.

Due to the difference in lines being used in each algorithm, newtons method is faster, but requires higher-order information (derivative) compared to the slower secant method that uses a secant line.

P.S. Evidently the bisection algorithm takes much longer (more iterations) to converge to the root compared to newtons or secant method.
