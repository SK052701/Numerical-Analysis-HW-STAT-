### q3. approximate $f(x)=e^x$ over $[-1,+1]$.

use chebyshev to approximate $f(x) = e^x$ over the interval $[-1,1]$ with number of data points $n = 11$ for both evenly spaced and chebyshev roots and for $n = 16, 21$ for chebyshev roots. give error bound for each case. plot $f(x)$ and all $p(x)$ on the same graph. dont forget to identify each $p(x)$ with its case. ie,

1. n = 11 data points (degree 10), evenly spaced;
2. n = 11 data points (degree 10), chebyshev roots;
3. n = 16 data points (degree 10), chebyshev roots; and
4. n = 21 data points (degree 10), chebyshev roots.


#### code template

```python
def all():
  # range [-1,+1]
  xa = -1.
  xb = 1.

  n, nn, nnn = 11, 16, 21 # number of data points that span range


  # BEGIN get polynomials FOR DEGREE n-1

  # set coefficients for standard monomial; ie, c = [1,1,1/2,...]
  c11_std = [1/math.factorial(i) for i in range(n+1)]
  p11_std = npp.Polynomial(c11_std) # standard monomial as class Polynomial

  c16_std = [1/math.factorial(i) for i in range(nn+1)]
  p16_std = npp.Polynomial(c16_std)

  c21_std = [1/math.factorial(i) for i in range(nnn+1)]
  p21_std = npp.Polynomial(c21_std)

  # let numpy convert to chebyshev polynomial of same degree
  p11_che = p11_std.convert(kind=npp.Chebyshev)
  c11_che = np.array(list(p11_che)) # fyi, chebychev coefficients

  p16_che = p16_std.convert(kind=npp.Chebyshev)
  c16_che = np.array(list(p16_che))

  p21_che = p21_std.convert(kind=npp.Chebyshev)
  c21_che = np.array(list(p21_che))


  # END get polynomials

  # BEGIN evaluate polynomials at resolution dx

  dx = 0.1 # x_delta = plot resolution
  iplot = int((xb-xa)/dx + 1) # points to plot at resolution dx

  ps = np.full((iplot,11),np.nan) # cols = x,f(x),em,[y,e]x[p_even,p_cheby]
  x = np.linspace(xa,xb,iplot); # space x evenly in [ax,xb]
  fx = np.exp(x) # f(x) = e^x ~ use system for "true value"
  ps[:,0] = x
  ps[:,1] = fx
  ps[:,2] = [np.finfo(float).eps for i in range(iplot)] # machine error ref

  # the usual polynomial derived from n evenly spaced points
  px11_std = np.polyval(c11_std,x)
  ps[:,3] = px11_std
  ps[:,4] =  fx - px11_std

  # chebyshev with n nodes
  px11_che = npp.chebyshev.chebval(x,c11_che)  # chebyshev polynomial as class
  ps[:,5] = px11_che
  ps[:,6] = fx - px11_che

  px16_che = npp.chebyshev.chebval(x,c16_che)  # chebyshev polynomial as class
  ps[:,7] = px16_che
  ps[:,8] = fx - px16_che

  px21_che = npp.chebyshev.chebval(x,c16_che)  # chebyshev polynomial as class
  ps[:,9] = px21_che
  ps[:,10] = fx - px21_che

  # END evaluate polynomials

  # BEGIN load dataframe

  # dataframe
  nCol = ["x","f(x)","em","p11_std","e11_std","p11_che","e11_che",'p16_che','e16_che','p21_che','e21_che']
  df = pd.DataFrame(data=ps,columns=nCol)
  if True:
    display(df)

  # END load dataframe

  # BEGIN plot

  # plots from dataframe
  if True:
    ax1  = df.plot(x = "x",y="f(x)",label="f(x)",color="0.9")
    ax11 = df.plot(x = "x",y="p11_std",label="p11,std",ax=ax1,color="mediumseagreen",marker=".",kind="scatter")
    ax12 = df.plot(x = "x",y="p11_che",label="p11,che",ax=ax1,color="xkcd:sky blue",marker="+",kind="scatter")
    ax13 = df.plot(x = "x",y="p16_che",label="p16,che",ax=ax1,color="orange",marker="*",kind="scatter")
    ax14 = df.plot(x = "x",y="p21_che",label="p21,che",ax=ax1,color="red",marker=">",kind="scatter")
    ax1.set_title("$e^x$")
    ax1.set_ylabel("x")
    ax1.set_ylabel("p(x)")

    ax2  = df.plot(x = "x",y="em",label="$\epsilon_M$",color="0.9")
    ax21 = df.plot(x = "x",y="e11_std",label="e11,std",ax=ax2,color="mediumseagreen",marker=".",kind="scatter")
    ax22 = df.plot(x = "x",y="e11_che",label="e11,che",ax=ax2,color="xkcd:sky blue",marker="+",kind="scatter")
    ax23 = df.plot(x = "x",y="e16_che",label="e16,che",ax=ax2,color="orange",marker="*",kind="scatter")
    ax24 = df.plot(x = "x",y="e21_che",label="e21,che",ax=ax2,color="red",marker=">",kind="scatter")
    ax2.set_title("error=f(x)-p(x)")
    ax2.set_ylabel("x")
    ax2.set_ylabel("e")

  # END plot

if __name__ == "__main__":
  all()
```

```python

e11_std_bound = 1.9 #from the above graph
e11_che_bound = math.factorial(9)/(math.factorial(11)*(2)**10) #from Sauer page 166
e16_che_bound = math.factorial(14)/(math.factorial(16)*(2)**15) #from Sauer page 166
e21_che_bound = math.factorial(19)/(math.factorial(21)*(2)**20) #from Sauer page 166
#since we are using a polynomial with coefficients [1, 1, 1/2, 1/3,...], the n'th derivative of this polynomial is (n-2)!

print(f'Error bound for standard polynomial interpolated on 11 evenly spaced points on [-1,1] (from the graph): {e11_std_bound}')
print(f'Error bound for chebyshev polynomial (standard polynomial interpolated at 11 chebyshev roots on [-1,1]): {e11_che_bound}')
print(f'Error bound for chebyshev polynomial (standard polynomial interpolated at 16 chebyshev roots on [-1,1]): {e16_che_bound}')
print(f'Error bound for chebyshev polynomial (standard polynomial interpolated at 21 chebyshev roots on [-1,1]): {e21_che_bound}')
```
