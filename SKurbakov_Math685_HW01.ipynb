```python
#conversion from decimal to binary (positive integers)

def dec_to_bin(integer):
  q = integer
  binary = ''
  while q > 0:
    q, r = (q//2, q%2)
    binary = binary + str(r)
  return binary[::-1]




print(dec_to_bin(55)) #110111
```

```python
#path of least resistance (all integers)

def deci_to_bina(decimal):
  return bin(decimal)[2:]




print(deci_to_bina(55)) #110111
```

```python
#conversion from decimal to binary (recursion)

def convert(decimal):

  if decimal > 0:
    q, r = (decimal//2, decimal%2)
    conv = convert(q)
    conv = conv + str(r)

  else:
    conv = ''

  return conv


print(convert(55)) #110111
```
