```python
#q2

#t0 = 1960
#t1 = 1970
#t3 = 1990
#t4 = 2000

t = [0,1,3,4] #time
y = [3039585530, 3707475887,5281653820, 6079603571] #population
ln_pop = np.array(np.log(y)) #natural log of the population

matx = np.array([[1, year] for year in t]) # A
b = ln_pop #b



matx_trans = matx.T #transpose of A

left = matx_trans @ matx #left hand side of the orthogonal equation
right = matx_trans @ b #right hand side of the orthogonal equation

k,r2 = np.linalg.solve(left, right) #coefficients of the linearized exp model

r1 = np.exp(k) #coefficient of the exp model

print(f'The exponential model with smallest error achieved through linearization: {r1}*e^({r2}*t)')

LSE = 0 #least square error

for tm, p in zip(t,ln_pop):
  LSE+= (k+r2*tm-p)**2

print(f'The least square error of the linearized exponential model is: {LSE}')
```
